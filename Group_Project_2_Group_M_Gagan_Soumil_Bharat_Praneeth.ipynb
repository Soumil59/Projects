```python
#problem7

housing_data = pd.read_csv("housing.csv", usecols=["District Name", "Rural/Urban",
                                                         "Total Number of households",
                                                         "Total Number of Livable",
                                                         "Total Number of Dilapidated",
                                                         "Latrine_premise"])



housing_data["Total Number of Dilapidated"] = housing_data.apply(
    lambda row: row["Total Number of Dilapidated"] / 100 * row["Total Number of households"],
    axis=1
)

housing_data["Latrine_premise"] = housing_data.apply(
    lambda row: row["Latrine_premise"] / 100 * row["Total Number of households"],
    axis=1
)

housing_data["Total Number of Dilapidated (Rural)"] = housing_data.apply(
    lambda row: row["Total Number of Dilapidated"] if row["Rural/Urban"] == "Rural" else 0,
    axis=1
)

housing_data["Total Number of Dilapidated (Urban)"] = housing_data.apply(
    lambda row: row["Total Number of Dilapidated"] if row["Rural/Urban"] == "Urban" else 0,
    axis=1
)

housing_data["Latrine_premise (Rural)"] = housing_data.apply(
    lambda row: row["Latrine_premise"] if row["Rural/Urban"] == "Rural" else 0,
    axis=1
)

housing_data["Latrine_premise (Urban)"] = housing_data.apply(
    lambda row: row["Latrine_premise"] if row["Rural/Urban"] == "Urban" else 0,
    axis=1
)

print(housing_data)
```

```python
#Problem8
import matplotlib.pyplot as plt

states = ['State A', 'State B', 'State C', 'State D']
households_per_100_people = [75, 85, 90, 80]
toilet_percentage = [70, 75, 80, 85]
urban_population = [60, 40, 70, 50]
rural_population = [40, 60, 30, 50]

plt.figure(figsize=(10, 5))
plt.bar(states, households_per_100_people, color='skyblue')
plt.title('Number of Households for 100 People by State')
plt.xlabel('State')
plt.ylabel('Households per 100 People')
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(6, 6))
plt.pie(toilet_percentage, labels=states, autopct='%1.1f%%', colors=['lightgreen', 'lightcoral', 'lightskyblue', 'lightpink'])
plt.title('Percentage of Households with Toilet(s) in Premise by State')
plt.show()

plt.figure(figsize=(8, 6))
plt.scatter(urban_population, rural_population, s=100, c='orange', alpha=0.7)
plt.title('Urban to Rural Population Ratio by State')
plt.xlabel('Urban Population')
plt.ylabel('Rural Population')
for i, txt in enumerate(states):
    plt.annotate(txt, (urban_population[i], rural_population[i]), xytext=(5, -5), textcoords='offset points')
plt.grid(True)
plt.show()
```
